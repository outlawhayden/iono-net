```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def convert_to_complex(s):
    if s == "NaNNaNi":
        return complex(np.nan)
    else:
        return complex(s.replace('i', 'j'))


sample_df = pd.read_csv("/home/houtlaw/iono-net/data/baselines/10k_lownoise/train_image_recon_jnp_unfocused.csv", dtype=str).map(convert_to_complex).to_numpy().T

SAMPLE_IDX = 33

sample_df
```
